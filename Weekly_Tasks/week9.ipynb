```python
import matplotlib.pyplot as plt

with open("../Data/cookiepath.txt") as file:
    path = file.read()

s_path = [[float(xy.split(' ')[0]), float(xy.split(' ')[1])] for xy in path.split('\n') if xy != '']
hs = [x[0] for x in s_path]
cs = [y[1] for y in s_path]
x = [i/5 for i in range(len(hs))]

plt.title('Huber cookie position')
plt.plot(x, hs, color='blue', label='Hubert')
plt.plot(x, cs, color='orange', label='cookie')
plt.show()
```

Hubert moves nicely to the next cookie, surprisingly...
