```python
import sys
import matplotlib.pyplot as plt
import numpy as np
import copy
```

```python
with open("../Data/NanoGigaVIZ.txt") as file:
    path = file.read()

s_path = [[float(xy.split(' ')[0]), float(xy.split(' ')[1])] for xy in path.split('\n') if xy != '']


fig, ax = plt.subplots()
circles = [(11, 11), (5, 5), (4, 5), (6, 5), (6, 2), (2, 9), (6, 12), (8, 10), (11, 2), (11, 3), (11, 7)]
ax.add_patch(plt.Circle(circles[0], 1, color='green'))
for circle in circles[1:]:
    ax.add_patch(plt.Circle(circle, 1, color='black'))
lastX, lastY = 2, 2
for x, y in s_path:
    ax.plot([lastX, x], [lastY, y], color="blue")
    lastY = y
    lastX = x
plt.show()
```
